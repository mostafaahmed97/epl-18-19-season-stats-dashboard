```python
matches_df = pd.read_csv('./data/epl-matches-18-19.csv')
```

```python
matches_df.columns
```

```python
teams = matches_df["home_team_name"].unique()
cum_season_stats = {}

home_stats_seletor = {
    "home_team_goal_count": "goals",
    "home_team_yellow_cards": "yellow_cards",
    "home_team_red_cards": "red_cards",
    "home_team_fouls": "fouls",
}

away_stats_seletor = {
    "away_team_goal_count": "goals",
    "away_team_yellow_cards": "yellow_cards",
    "away_team_red_cards": "red_cards",
    "away_team_fouls": "fouls",
}

stats = ["timestamp", "date_GMT", "goals", "yellow_cards", "red_cards", "fouls"]
cumsum_stats = ['goals', 'yellow_cards','red_cards', 'fouls']

for team in teams:
    agg_stats = pd.DataFrame()

    home_matches = matches_df[matches_df["home_team_name"] == team]
    away_matches = matches_df[matches_df["away_team_name"] == team]

    agg_stats = agg_stats.append(home_matches.rename(columns=home_stats_seletor)[stats])
    agg_stats = agg_stats.append(away_matches.rename(columns=away_stats_seletor)[stats])
    agg_stats = agg_stats.sort_index(inplace=False)
    
    summed_stats = pd.DataFrame()
    summed_stats[cumsum_stats] = agg_stats[cumsum_stats].cumsum()

    cum_season_stats[team] = summed_stats.to_dict(orient='list')


print(cum_season_stats)
```

```python
import json

with open('./data/cumulative-season-stast.json', 'w') as f:
    json.dump(cum_season_stats, f)
```
